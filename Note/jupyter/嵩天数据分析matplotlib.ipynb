Matplotlib库由各种可视化类构成，内部结构复杂，受Matlab启发

```python
import matplotlib.pyplot as plt
plt.plot([12,3,4,56,78,7,9,0])
plt.ylabel("number")
plt.show()
```

> plt.plot()只有一个输入列表或数组时，参数被当作Y轴，X轴以索引自动生成

```python
import matplotlib.pyplot as plt
plt.plot([12,3,4,56,78,7,9,0],[-2,-1,0,1,2,3,4,5])
plt.ylabel("number")
plt.show()
```

> plt.plot(x,y)当有两个以上参数时，按照X轴和Y轴顺序绘制数据点

```python
import matplotlib.pyplot as plt
plt.plot([12,3,4,56,78,7,9,0],[-2,-1,0,1,2,3,4,5])
plt.ylabel("number")
plt.axis([-6,60,0,6])
plt.show()
```

> plt.axis([xa,xb,ya,yb]),其中参数数组前两个元素是x轴的起点和终点，后两个数组元素是y轴的起点和终点

## pyplot的绘图区域
`plt.subplot(nrows, ncols, plot_number)`
就是将一个画图区域分成好几部分，nrows,ncols,表示的意思是将一个画图区域分成nrows行ncols列，plot表示的意思是当前选中的画图区，比如`plt.subplot(2, 3, 4)`

|1|2|3|
|-|-|-|
|4,当前画图区域|5|6|

```python
import numpy as np
import matplotlib.pyplot as plt

#能量衰减函数
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

a=np.arange(0.0,5.0,0.01,dtype=np.float)
plt.subplot(2,1,1)
plt.plot(a,f(a))

plt.subplot(2,1,2)
plt.plot(a,np.cos(2*np.pi*a),"r--")

plt.show()
```

## pyplot的plot函数
`plt.plot(x,y, format_string, **kwargs)`
- x : X轴数据，列表list或数组ndarray，可选 
- y :Y轴数据，列表list或数组ndarray
- format_string: 控制曲线的格式字符串，可选
- **kwargs: 第二组或更多(x,y,format_string)，可以绘制多条曲线，但是每条曲线必须给定x，y，不能缺省

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.arange(10)
plt.plot(a,a*2,a,a*3,a,a*5)
```

> 为了区分不同的曲线，format_string参数是控制曲线的格式字符串，
有颜色字符，风格字符，标记字符组成，具体用到具体查询

## pyplot的中文显示
pyplot默认不支持中文，有两种方法显示中文
1. pyplot并不默认支持中文显示，需要rcParams修改字体实现，会修改所有字符的字体，英文也是

|属性|说明|
|-|-|
|'font.family'|用于显示字体的名字，'SimHei'中文黑体 'Kaiti'中文楷体 'LiSu'中文隶书 'FangSong'中文仿宋 'YouYuan'中文幼圆 'STSong'华文宋体
|
|'font.style'|字体风格，正常'normal'或斜体'italic'| 
|'font.size'|字体大小，整数字号或者'large'、'x‐small'|

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams["font.family"]="SimHei"#黑体
plt.plot([2,3,4,5,7])
plt.ylabel("中文测试")
plt.show()
```

2.  在有中文输出的地方，增加一个属性：fontproperties（推荐）


```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams["font.family"]="SimHei"#黑体
plt.plot([2,3,4,5,7])
plt.ylabel("中文测试",fontproperties="SimHei",fontsize=24)
plt.show()
```

## pyplot文本显示函数
|函数|说明|
|-|-|
|plt.xlabel()|对X轴增加文本标签|
|plt.ylabel()|对Y轴增加文本标签|
|plt.title()|对图形整体增加文本标签|
|plt.text()|在任意位置增加文本|
|plt.annotate()|在图形中增加带箭头的注解|

```python
import numpy as np
import matplotlib.pyplot as plt

a=np.arange(0,5,0.01)
plt.plot(a,np.cos(2*np.pi*a),"r--")

plt.xlabel("横轴：时间",fontsize=24)
plt.ylabel("纵轴：时间",fontsize=24)
plt.title("标题title",fontsize=24)
plt.annotate("我是一个箭头",xy=(2,0),xytext=(2,1),arrowprops=dict(facecolor="black",shrink=0.1,width=2))
```

## 饼图的绘制
`plt.pie()`

```python
import numpy as np
import matplotlib.pyplot as plt

labels="A","B","C","D"
sizes=np.array([20,40,10,30])
explode=(0,0.1,0,0)

plt.pie(sizes,explode=explode,labels=labels,autopct="%1.1f%%")
plt.show()
```

## pyplot直方图的绘制
`plt.hist()`
plt.hist()函数有很多参数 ，其中bins参数负责控制图表中绘制多少个直方的个数

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(20)
mu,sigma=1,3#均值和方差
a=np.random.normal(mu,sigma,size=100)

plt.subplot(1,2,1)
plt.hist(a,bins=10,histtype="stepfilled",facecolor="b",alpha=0.75)

plt.subplot(1,2,2)
plt.hist(a,bins=40,histtype="stepfilled",facecolor="b",alpha=0.75)
plt.show()
```
